## Loops


### Range


A veces, puede que quieras repetir una operación determinada muchas veces. Las ejecuciones repetidas como estas se realizan mediante **bucles**. Veremos dos tipos de bucles: los bucles `for` y los bucles `while`.

Antes de hablar de los bucles, analicemos el objeto `range`. Es útil pensar en el objeto `range` como una lista ordenada. Por ahora, veamos el caso más simple. Si queremos generar un objeto que contenga elementos ordenados del 0 al 2, simplemente usamos el siguiente comando:


```python
# Use the range

range(3)
```

### ¿Qué es un bucle `for`?


El bucle `for` te permite ejecutar un bloque de código múltiples veces. Por ejemplo, lo usarías si deseas imprimir cada elemento de una lista.  

Intentemos usar un bucle `for` para imprimir todos los años presentados en la lista `dates`:

Esto se puede hacer de la siguiente manera:

```python
# For loop example

dates = [1982,1980,1973]
N = len(dates)

for x in range(N):
    print(dates[x])     
```

El código que está indentado se ejecuta **N** veces; en cada ejecución, el valor de `i` se incrementa en 1.  
La instrucción que se ejecuta es `print`, la cual imprime el valor de la lista en el índice `i`, como se muestra a continuación:


<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/f3bJb12CNd053xJfk1wU2g/LoopsForRangee.gif" width="800">

En este ejemplo, podemos imprimir una secuencia de números del 0 al 7:

```python
# Example of for loop

for i in range(0, 8):
    print(i)
```

En Python, podemos acceder directamente a los elementos de una lista de la siguiente manera:


```python
# Exmaple of for loop, loop through list

for year in dates:  
    print(year)   
```

En cada iteración, el valor de la variable `year` se comporta como el valor de `dates[i]` en el primer ejemplo:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/LoopsForList.gif" width="800">

Podemos modificar los elementos de una lista:

```python
# Use for loop to change the elements in list

squares = ['red', 'yellow', 'green', 'purple', 'blue']

for i in range(0, 5):
    print("Before square ", i, 'is',  squares[i])
    squares[i] = 'white'
    print("After square ", i, 'is',  squares[i])
```

Podemos acceder al índice y a los elementos de una lista de la siguiente manera:


```python
# Loop through the list and iterate on both index and element value

squares=['red', 'yellow', 'green', 'purple', 'blue']

for i, square in enumerate(squares):
    print(i, square)
```

### What is <code>while</code> loop?

Como puedes ver, el bucle `for` se utiliza para un flujo de repetición controlado. Sin embargo, ¿qué pasa si no sabemos cuándo queremos detener el bucle? ¿Qué ocurre si queremos seguir ejecutando un bloque de código hasta que se cumpla una condición determinada?

El bucle `while` existe como una herramienta para la ejecución repetida basada en una condición. El bloque de código continuará ejecutándose hasta que la condición lógica proporcionada devuelva un valor booleano **False**.
