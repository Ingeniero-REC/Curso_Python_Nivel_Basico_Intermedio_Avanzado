<h2 id="tuple">Tuplas</h2>

En Python, existen diferentes tipos de datos, como **String (cadena de texto)**, **Integer (entero)** y **Float (número de punto flotante)**. Todos estos tipos de datos pueden almacenarse dentro de una tupla de la siguiente manera:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesType.png" width="550" align="center">


Ahora, creemos tu primera tupla con una cadena de texto (*string*), un número entero (*integer*) y un número de punto flotante (*float*).

````python
# Crea tu primera tupla
tuple1 = ("disco",10, 1.2)
print(tuple1)
````

El tipo de la variable es una **tupla**.

````python
# Imprime el tipo de tupla que creaste
type(tuple1)
````

✔️ ¿Por qué usar tuplas?

Son más rápidas que las listas.

Se usan para datos que no deben cambiar.

Útiles para retornar múltiples valores desde una función

<h3 id="index">Indexación</h3>

Cada elemento de una tupla puede accederse mediante un **índice**. La siguiente tabla representa la relación entre el índice y los elementos de la tupla. Cada elemento puede obtenerse utilizando el nombre de la tupla seguido de corchetes (`[]`) que contienen el número del índice:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesIndex.gif" width="450" align="center">

Podemos imprimir cada valor de la tupla:

````python
# Imprimir la variable en cada índice
print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
````

Podemos imprimir el **tipo** de cada valor en la tupla:

````python
# Imprimir el tipo de cada dato de la tupla
print(type(tuple1[0]))
print(type(tuple1[1]))
print(type(tuple1[2]))
````

También podemos utilizar **índices negativos**. Usaremos la misma tabla anterior, pero con los índices negativos correspondientes:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesNeg.png" width="450" align="center">


Podemos obtener el último elemento de la siguiente manera (esta vez no usaremos la instrucción print para mostrar los valores):

````python
# Utilice un índice negativo para obtener el valor del último elemento
tuple1[-1]
````

Podemos mostrar los siguientes dos elementos de la siguiente manera:

````python
# Utilice un índice negativo para obtener el valor del segundo último elemento
tuple1[-2]
````

````python
# Utilice un índice negativo para obtener el valor del tercer último elemento
tuple1[-3]
````

<h3 id="concate">Concatenar tuplas</h3>

Podemos **concatenar** o **combinar** tuplas utilizando el signo **+**:

````python
# Concatenar dos tuplas
tuple2 = tuple1 + ("hard rock", 10)
tuple2
````

Podemos **segmentar (hacer *slicing* de)** las tuplas para obtener varios valores, como se muestra en la siguiente figura:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesSlice.gif" width="450" align="center">

<h3 id="slice">Slicing</h3>

Podemos cortar tuplas, obteniendo nuevas tuplas con los elementos correspondientes:

````python
# Rebanada del índice 0 al índice 2
tuple2[0:3]
````

Podemos obtener los dos últimos elementos de la tupla:

````python
# Rebanada del índice 3 al índice 4
tuple2[3:5]
````

Podemos obtener la longitud de una tupla usando el comando length:

````python
# Obtener la longitud de la tupla
len(tuple2)
````

Esta figura muestra el número de elementos:

<h3 id="sort">Ordenamiento</h3>

Considere la siguiente tupla:

````python
# Una tupla de muestra
Ratings = (0, 9, 6, 5, 10, 8, 9, 6, 2)
````

Podemos ordenar los valores de una tupla y guardarla en un nuevo objeto. El resultado es una nueva lista ordenada.

````python
# Ordenar la tupla
RatingsSorted = sorted(Ratings)
RatingsSorted
````

<h3 id="nest">Tupla anidada</h3>

Una tupla puede contener otra tupla, así como otros tipos de datos más complejos. Este proceso se denomina "anidación". Considere la siguiente tupla con varios elementos:

````python
# Crear una tupla de nido
NestedT =(1, 2, ("pop", "rock") ,(3,4),("disco",(1,2)))
````

Cada elemento de la tupla, incluidas otras tuplas, se puede obtener a través de un índice como se muestra en la figura:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesNestOne.png" width="550" align="center">

````python
#Imprimir elemento en cada índice
print("Element 0 of Tuple: ", NestedT[0])
print("Element 1 of Tuple: ", NestedT[1])
print("Element 2 of Tuple: ", NestedT[2])
print("Element 3 of Tuple: ", NestedT[3])
print("Element 4 of Tuple: ", NestedT[4])
````

Podemos utilizar el segundo índice para acceder a otras tuplas como se muestra en la figura:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesNestTwo.png" width="550" align="center">

Nosotros podemos acceder a las tuplas anidadas:

````python
# Imprimir elemento en cada índice, incluidos los índices anidados
print("Element 2, 0 of Tuple: ",   NestedT[2][0])
print("Element 2, 1 of Tuple: ",   NestedT[2][1])
print("Element 3, 0 of Tuple: ",   NestedT[3][0])
print("Element 3, 1 of Tuple: ",   NestedT[3][1])
print("Element 4, 0 of Tuple: ",   NestedT[4][0])
print("Element 4, 1 of Tuple: ",   NestedT[4][1])
````

Podemos acceder a las cadenas en las segundas tuplas anidadas usando un tercer índice:

````python
# Imprime el primer elemento en la segunda tupla anidada
NestedT[2][1][0]
````

````python
# Imprime el segundo elemento en la segunda tupla anidada
NestedT[2][1][1]
````

Podemos usar un árbol para visualizar el proceso. Cada nuevo índice corresponde a un nivel más profundo del árbol:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesNestThree.gif" width="550" align="center">

De manera similar, podemos acceder a elementos anidados más profundamente en el árbol con un tercer índice:

````python
# Imprime el primer elemento en la segunda tupla anidada
NestedT[4][1][0]
````

````python
# Imprime el segundo elemento en la segunda tupla anidada
NestedT[4][1][1]
````

La siguiente figura muestra la relación del árbol y el elemento <code>NestedT\[4]\[1]\[1]</code>:
