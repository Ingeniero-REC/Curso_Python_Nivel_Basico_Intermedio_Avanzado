# Secuencias de escape

Las barras invertidas \ indican el inicio de una secuencia de escape. Las secuencias de escape representan caracteres especiales que pueden ser difíciles de escribir directamente. Por ejemplo, \n representa un salto de línea. Cuando Python encuentra \n, el texto que sigue se muestra en una nueva línea:

````python
# Secuencia de escape de nueva línea
print("The BodyGuard \nes el mejor album de musica")
````

De manera similar, la barra invertida seguida de "t" (\t) representa una tabulación.

````python
# Secuencia de escape de tabulación
print("The BodyGuard \tes el mejor album de musica")
````

Si quieres colocar una barra invertida \ en tu cadena, utiliza una doble barra invertida \ \ .

````python
# Incluir barra invertida en la cadena
print("The BodyGuard \\ es el mejor album de musica")
````

Podemos también colocar una "r" antes de la cadena para mostrar la barra invertida:

````python
# r le indicará a Python que la cadena se mostrará como una cadena sin procesar (raw string)
print(r"The BodyGuard \nes el mejor album de musica")
````

# Operaciones de manipulación de cadenas

Existen muchos métodos de cadena en Python que se pueden usar para manipular datos. Vamos a usar algunas operaciones básicas de cadenas en los datos.

<p>Probemos el método <code>upper</code>; este método convierte los caracteres en minúsculas a caracteres en mayúsculas:</p>

````python
# Convertir todos los caracteres de la cadena a mayúsculas

a = "Thriller is the sixth studio album"
print("before upper:", a)
b = a.upper()
print("After upper:", b)
````

````python
# Convertir todos los caracteres de la cadena a minusculas
a = "THRILLER IS THE SIXTH STUDIO ALBUM"
print("before lower:", a)
b = a.lower()
print("After lower:", b)
````

El metodo <code>replace</code> reemplaza un segmento de la cadena, es decir, una subcadena, por una cadena nueva. Introducimos la parte de la cadena que queremos cambiar. El segundo argumento es aquello por lo que queremos intercambiar el segmento, y el resultado es una nueva cadena con el segmento modificado:

# RegEx

In Python, **RegEx** (abreviatura de **Regular Expression**, o **Expresión Regular**) es una herramienta utilizada para buscar, hacer coincidir y manipular cadenas de texto.

Este módulo de **RegEx** proporciona varias funciones para trabajar con expresiones regulares, entre ellas <code>search</code>, <code>split</code>, <code>findall</code> y <code>sub</code>.

Python proporciona un módulo integrado llamado <code>re</code>, que permite trabajar con expresiones regulares.

Primero, importa el módulo <code>re</code>:

````python
import re
````

La función <code>search()</code> busca patrones específicos dentro de una cadena. A continuación, se muestra un ejemplo que explica cómo utilizar la función <code>search()</code> para buscar la palabra <code>"Body"</code> en la cadena <code>"The BodyGuard is the best"</code>.

````python
s1 = "The BodyGuard es el mejor album"

# Definir el patrón a buscar
patron = r"Body"

# Usar la función search() para buscar el patrón en la cadena
buscarPatron = re.search(patron,s1)

# Verificar si se encontró una coincidencia
if buscarPatron:
    print("Coincidencia encontrada")
else:
    print("Patron no encontrado")
````

Las **expresiones regulares (RegEx)** son patrones utilizados para buscar y manipular cadenas de texto. Existen varias secuencias especiales en RegEx que permiten hacer coincidir caracteres o patrones específicos.

| Secuencia especial | Significado                                                                               | Ejemplo                                                                              |
| ------------------ | ----------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------ |
| `\d`               | Coincide con cualquier dígito (0-9)                                                       | `"123"` coincide con `\d\d\d`                                                        |
| `\D`               | Coincide con cualquier carácter que **no** sea un dígito                                  | `"hello"` coincide con `\D\D\D\D\D`                                                  |
| `\w`               | Coincide con cualquier carácter de palabra (a-z, A-Z, 0-9 y _)                            | `"hello_world"` coincide con `\w\w\w\w\w\w\w\w\w\w\w`                                |
| `\W`               | Coincide con cualquier carácter que **no** sea de palabra                                 | `"@#$%"` coincide con `\W\W\W\W`                                                     |
| `\s`               | Coincide con cualquier carácter de espacio en blanco (espacio, tab, salto de línea, etc.) | `"hello world"` coincide con `\w\w\w\w\w\s\w\w\w\w\w`                                |
| `\S`               | Coincide con cualquier carácter que **no** sea espacio en blanco                          | `"hello_world"` coincide con `\S\S\S\S\S\S\S\S\S\S\S`                                |
| `\b`               | Coincide en el límite entre un carácter de palabra y uno que no lo es                     | `"cat"` coincide con `\bcat\b` en `"The cat sat on the mat"`                         |
| `\B`               | Coincide en cualquier posición que **no** sea un límite de palabra                        | `"cat"` coincide con `\Bcat\B` en `"category"` pero no en `"The cat sat on the mat"` |


Ejemplos de secuencias especiales:

Un ejemplo sencillo de cómo usar la secuencia especial \d en un patrón de expresión regular con código Python:

````python
import re

patron = r"\d\d\d\d\d\d\d\d\d\d"  # Coincide con cualquier secuencia de diez dígitos consecutivos
texto = "Mi número de teléfono es 1234567890"
coincidencia = re.search(patron, texto)

if coincidencia:
    print("Número de teléfono encontrado:", coincidencia.group())
else:
    print("No se encontró coincidencia")
````

El método <code>match.group()</code> se utiliza en el módulo <code>re</code> de Python para recuperar la parte de la cadena donde el patrón de la expresión regular encontró una coincidencia. A continuación, se presenta una explicación detallada:

**Propósito**
- **Extraer el texto coincidente:** <code>match.group()</code> devuelve exactamente la subcadena que coincide con el patrón.

**Uso**
- Cuando utilizas funciones como <code>re.search()</code> o <code>re.match()</code>, estas devuelven un objeto de coincidencia (*match object*) si se encuentra el patrón. Luego, puedes usar <code>match.group()</code> para obtener el texto que coincidió con dicho patrón.

En este caso, <code>match.group()</code> recupera la subcadena <code>1234567890</code> del texto, que es la parte que coincide con el patrón.

El patrón de la expresión regular se define como <code>r"\d\d\d\d\d\d\d\d\d\d"</code>, el cual utiliza la secuencia especial <code>\d</code> para coincidir con cualquier dígito (0-9). La secuencia <code>\d</code> se repite diez veces para hacer coincidir diez dígitos consecutivos.

Un ejemplo sencillo de cómo utilizar la secuencia especial <code>\W</code> en un patrón de expresión regular con código en Python:

````python
# Coincide con cualquier carácter que no sea alfanumérico
import re

patron = r"\W"  # Coincide con cualquier carácter que NO sea alfanumérico ni guion bajo
texto = "¡Hola, mundo!"
coincidencias = re.findall(patron, texto)

print("Coincidencias:", coincidencias)
````

El patrón de la expresión regular se define como <code>r"\W"</code>, el cual utiliza la secuencia especial <code>\W</code> para coincidir con cualquier carácter que **no** sea un carácter de palabra (a-z, A-Z, 0-9 o <code>_</code>). La cadena en la que se buscarán las coincidencias es <code>"Hello, world!"</code>.

La función <code>findall()</code> encuentra todas las ocurrencias de un patrón específico dentro de una cadena.

````python
import re

s2 = "The BodyGuard is the best album of 'Whitney Houston'."

# Usar la función findall() para encontrar todas las apariciones de "st" en la cadena
resultado = re.findall("st", s2)

# Imprimir la lista de coincidencias encontradas
print(resultado)
````

La función <code>split()</code> de las expresiones regulares divide una cadena en una lista de subcadenas utilizando un patrón específico como separador.

````python
import re

s2 = "The BodyGuard is the best album of 'Whitney Houston'."

# Usar la función split() para dividir la cadena usando "\s" (espacios en blanco)
arreglo_dividido = re.split(r"\s", s2)

# El arreglo_dividido contiene todas las subcadenas, separadas por caracteres de espacio
print(arreglo_dividido)
````

A continuación, se presenta una explicación detallada:

<code>re.split(r"\s", s2)</code>:

**<code>re.split()</code>**: Esta función divide una cadena cada vez que encuentra una coincidencia con el patrón especificado.

- **<code>r"\s"</code>**: Es un patrón de expresión regular que coincide con cualquier carácter de espacio en blanco (espacios, tabulaciones, saltos de línea, etc.).
- **<code>s2</code>**: Es la cadena que deseas dividir.

La función <code>sub()</code> de las expresiones regulares en Python se utiliza para reemplazar todas las ocurrencias de un patrón en una cadena por un valor de reemplazo especificado.

````python
import re

s2 = "The BodyGuard is the best album of 'Whitney Houston'."

# Definir el patrón de expresión regular a buscar
patron = r"Whitney Houston"

# Definir la cadena de reemplazo
reemplazo = "leyenda"

# Usar la función sub() para reemplazar el patrón por la cadena de reemplazo
nueva_cadena = re.sub(
    patron,
    reemplazo,
    s2,
    flags=re.IGNORECASE
)  # re.IGNORECASE hace que la búsqueda no distinga entre mayúsculas y minúsculas

# La nueva_cadena contiene la cadena original con el patrón reemplazado
print(nueva_cadena)
````
