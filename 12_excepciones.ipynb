## ¿Qué es una excepción?

En esta sección aprenderás qué es una excepción y verás ejemplos de ellas.

### Definición

Una excepción es un error que ocurre durante la ejecución del código. Este error provoca que el código genere una excepción y, si no está preparado para manejarla, detendrá la ejecución del código.

### Ejemplos

Ejecuta cada fragmento de código y observa la excepción generada.

````python
1/0
````

<code>ZeroDivisionError</code> ocurre cuando intentas dividir entre cero.

````python
y = a + 5
````

<code>NameError</code> -- en este caso, significa que intentaste utilizar la variable <code>a</code> cuando esta no había sido definida.

````python
a = [1, 2, 3]
a[10]
````

<code>IndexError</code> -- en este caso, ocurrió porque intentaste acceder a un dato de una lista utilizando un índice que no existe para esa lista.

Existen muchas más excepciones integradas en Python, aquí tienes una lista de ellas:
https://docs.python.org/3/library/exceptions.html

## Manejo de excepciones

En esta sección aprenderás cómo manejar excepciones. Comprenderás cómo hacer que tu programa ejecute tareas específicas en lugar de detener la ejecución del código cuando se encuentra una excepción.

### Try Except

Un bloque <code>try except</code> te permitirá ejecutar código que podría generar una excepción y, en caso de que ocurra cualquier excepción o una específica, podremos manejar o capturar la excepción y ejecutar un código determinado. Esto nos permitirá continuar la ejecución de nuestro programa incluso si ocurre una excepción.

Python intenta ejecutar el código dentro del bloque <code>try</code>. En este caso, si el código del bloque <code>try</code> genera alguna excepción, esta será capturada y se ejecutará el bloque de código dentro del bloque <code>except</code>. Después de eso, se ejecutará el código que viene <em>después</em> del bloque try except.

````python
# Código potencial antes del bloque try except

try:
    # código que se intentará ejecutar
except:
    # código que se ejecutará si ocurre una excepción
    
# código que seguirá ejecutándose si ocurre una excepción
````

 ### Ejemplo de Try Except

En este ejemplo intentamos dividir un número proporcionado por el usuario, guardar el resultado en la variable <code>a</code> y luego mostrar el resultado de la operación. Al recibir una entrada del usuario y dividir un número entre ella, pueden generarse algunas excepciones. Por ejemplo, si intentamos dividir entre cero.

Intenta ejecutar el siguiente bloque de código con <code>b</code> como un número. Solo se generará una excepción si <code>b</code> es cero.

````python
a = 1

try:
    b = int(input("Please enter a number to divide a"))
    a = a / b
    print("Success a=", a)

except:
    print("There was an error")
````

### Try Except específico

Una instrucción específica de <code>try except</code> te permite capturar ciertas excepciones y también ejecutar determinado código dependiendo de la excepción. Esto es útil si no quieres manejar algunas excepciones y la ejecución debería detenerse. También puede ayudarte a encontrar errores en tu código de los que quizás no seas consciente. Además, puede ayudarte a diferenciar las respuestas para diferentes excepciones. En este caso, el código después del bloque try except podría no ejecutarse dependiendo del error.

<b>No ejecutar, solo para ilustrar:</b>

````python
# código potencial antes del try catch

try:
    # código que se intenta ejecutar
except (ZeroDivisionError, NameError):
    # código que se ejecuta si hay una excepción de los tipos indicados
    
# código que se ejecutará si no hay una excepción o si es una que estamos manejando
````

````python
# código potencial antes del try catch

try:
    # código que se intenta ejecutar
except ZeroDivisionError:
    # código que se ejecuta si hay un ZeroDivisionError
except NameError:
    # código que se ejecuta si hay un NameError
    
# código que se ejecutará si no hay una excepción o si es una que estamos manejando
````

También puedes tener un <code>except</code> vacío al final para capturar una excepción inesperada:

````python
# código potencial antes del try catch

try:
    # código que se intenta ejecutar
except ZeroDivisionError:
    # código que se ejecuta si hay un ZeroDivisionError
except NameError:
    # código que se ejecuta si hay un NameError
except:
    # código que se ejecuta si hay cualquier excepción
    
# código que se ejecutará si no hay una excepción o si es una que estamos manejando
````

### Ejemplo específico de Try Except

Este es el mismo ejemplo de arriba, pero ahora agregaremos mensajes diferenciados dependiendo de la excepción, informándole al usuario qué está mal con la entrada.

````python
a = 1

try:
    b = int(input("Por favor ingresa un número para dividir a"))
    a = a/b
    print("Éxito a=", a)
except ZeroDivisionError:
    print("El número que proporcionaste no puede dividir a 1 porque es 0")
except ValueError:
    print("No proporcionaste un número")
except:
    print("Algo salió mal")
````

### Try Except Else y Finally

<code>else</code> permite comprobar si no hubo ninguna excepción al ejecutar el bloque <code>try</code>. Esto es útil cuando queremos ejecutar algo solamente si no ocurrieron errores.

<b>no ejecutar, solo para ilustrar</b>

````python
# código potencial antes del try catch

try:
    # código que se intenta ejecutar
except ZeroDivisionError:
    # código que se ejecuta si hay un ZeroDivisionError
except NameError:
    # código que se ejecuta si hay un NameError
except:
    # código que se ejecuta si hay cualquier excepción
else:
    # código que se ejecuta si no hay ninguna excepción
    
# código que se ejecutará si no hay una excepción o si es una que estamos manejando
````

<code>finally</code> nos permite ejecutar siempre algo, haya ocurrido una excepción o no. Normalmente se utiliza para indicar el final del bloque <code>try except</code>.

````python
# código potencial antes del try catch

try:
    # código que se intenta ejecutar
except ZeroDivisionError:
    # código que se ejecuta si hay un ZeroDivisionError
except NameError:
    # código que se ejecuta si hay un NameError
except:
    # código que se ejecuta si hay cualquier excepción
else:
    # código que se ejecuta si no hay ninguna excepción
finally:
    # código que se ejecuta al final del try except sin importar lo que ocurra
    
# código que se ejecutará si no hay una excepción o si es una que estamos manejando
````

### Ejemplo de Try Except Else y Finally

Quizás hayas notado que, incluso si hay un error, el valor de <code>a</code> siempre se imprime. Usemos <code>else</code> e imprimamos el valor de <code>a</code> solo si no hay ningún error.

````python
a = 1

try:
    b = int(input("Por favor ingresa un número para dividir a"))
    a = a/b
except ZeroDivisionError:
    print("El número que proporcionaste no puede dividir a 1 porque es 0")
except ValueError:
    print("No proporcionaste un número")
except:
    print("Algo salió mal")
else:
    print("Éxito a=", a)
````

Ahora informemos al usuario que hemos terminado de procesar su respuesta. Usando <code>finally</code>, agreguemos un mensaje de impresión.

````python
a = 1

try:
    b = int(input("Por favor ingresa un número para dividir a"))
    a = a/b
except ZeroDivisionError:
    print("El número que proporcionaste no puede dividir a 1 porque es 0")
except ValueError:
    print("No proporcionaste un número")
except:
    print("Algo salió mal")
else:
    print("Éxito a=", a)
finally:
    print("Procesamiento completado")
````

## Ejercicio 1: Manejo de ZeroDivisionError

Imagina que tienes dos números y quieres determinar qué sucede cuando divides un número entre otro. Para hacer esto, necesitas crear una función de Python llamada `safe_divide.` Le proporcionas a esta función dos números, un `'numerator'` y un `'denominator'`. El `'numerator'` es el número que quieres dividir, y el `'denominator'` es el número entre el cual quieres dividir. Usa el método de entrada de usuario (`input`) de Python para obtener los valores.

La función debe ser capaz de realizar la división por ti y darte el resultado. Pero aquí está la condición: si intentas dividir entre cero (lo cual no está permitido en matemáticas), la función debe ser lo suficientemente inteligente como para detectar ese caso e indicarte que no es posible dividir entre cero. En lugar de mostrar un error, debe devolver `None`, lo que significa "nada" o "sin valor", y mostrar el mensaje `"Error: Cannot divide by Zero."`

````python
# Type your code here
def safe_divide(numerator, denominator):
    try:
        result = numerator / denominator
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by Zero.")
        return None

# Test case
numerator = int(input("Enter the numerator value:- "))
denominator = int(input("Enter the denominator value:- "))

print(safe_divide(numerator, denominator))
````

## Ejercicio 3: Manejo de excepciones genéricas

Imagina que tienes un número y quieres realizar una tarea matemática compleja. El cálculo requiere dividir el valor del argumento de entrada `"num"` entre la diferencia entre `"num"` y 5, y el resultado debe almacenarse en una variable llamada `"result"`.

Debes definir una función para que pueda realizar esa tarea matemática compleja. La función debe manejar cualquier posible error que ocurra durante el cálculo. Para hacerlo, puedes utilizar un bloque `try-except`. Si ocurre alguna excepción durante el cálculo, debe capturar el error utilizando la clase de excepción genérica `"Exception" as "e"`.

Cuando ocurra una excepción, la función debe mostrar el mensaje `"An error occurred during calculation.`

````python
# Type your code here
def complex_calculation(num):
    try:
        result = num / (num - 5)
        print(f"Result: {result}")
    except Exception as e:
        print("An error occurred during calculation.")
        print(e)

# Test case
user_input = float(input("Enter a number: "))
complex_calculation(user_input)
````
