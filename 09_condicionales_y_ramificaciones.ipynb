### Operadores de comparación

Las operaciones de comparación comparan un valor u operando y, según una condición, generan un valor booleano. Al comparar dos valores, se pueden usar estos operadores:

<ul>
<li>igual: <b>==</b></li>
<li>distinto: <b>!=</b></li>
<li>mayor que: <b>></b></li>
<li>menor que: <b>&lt;</b></li>
<li>mayor o igual que: <b>>=</b></li>
<li>menor o igual que: <b>&lt;=</b></li>
</ul>

Asignemos a <code>a</code> el valor 5. Utilice el operador de igualdad, denotado con dos signos de igual <b>==</b>, para determinar si dos valores son iguales. El siguiente caso compara la variable <code>a</code> con 6.

```python
# Condition Equal

a = 5
a == 6
```

El resultado es <b>Falso</b>, ya que 5 no es igual a 6.

Considere el siguiente operador de comparación de igualdad: <code>i > 5</code>. Si el valor del operando izquierdo, en este caso la variable <b>i</b>, es mayor que el valor del operando derecho, en este caso 5, la afirmación es <b>Verdadera</b>. En caso contrario, es <b>Falsa</b>. Si <b>i</b> es igual a 6, porque 6 es mayor que 5, la salida es <b>Verdadera</b>.

```python
# Greater than Sign

i = 6
i > 5
```

Establezca <code>i = 2</code>. La afirmación es falsa porque 2 no es mayor que 5:

```python
# Greater than Sign

i = 2
i > 5
```

Mostremos algunos valores para <code>i</code> en la figura. Establezca los valores mayores a 5 en verde y el resto en rojo. La región verde representa dónde la condición es **Verdadera**, y la roja, dónde la afirmación es **Falsa**. Si el valor de <code>i</code> es 2, obtenemos **Falso**, ya que el 2 se encuentra en la región roja. De igual manera, si el valor de <code>i</code> es 6, obtenemos **Verdadero**, ya que la condición se encuentra en la región verde.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsGreater.gif" width="650">

La prueba de desigualdad utiliza un signo de exclamación antes del signo igual. Si dos operandos no son iguales, la condición se convierte en **Verdadera**. Por ejemplo, la siguiente condición será **Verdadera** siempre que el valor de <code>i</code> no sea igual a 6:

```python
# Inequality Sign

i = 2
i != 6
```

Cuando <code>i</code> es igual a 6 la expresión de desigualdad produce <b>Falso</b>.

```python
# Inequality Sign

i = 6
i != 6
```

Vea la recta numérica a continuación. Cuando la condición es **Verdadera**, los números correspondientes se marcan en verde, y si la condición es **Falsa**, el número correspondiente se marca en rojo. Si igualamos <code>i</code> a 2, el operador es verdadero, ya que 2 está en la región verde. Si igualamos <code>i</code> a 6, obtenemos un **Falso**, ya que la condición está en la región roja.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsIneq.gif" width="650">


Podemos aplicar los mismos métodos a cadenas. Por ejemplo, podemos usar un operador de igualdad en dos cadenas diferentes. Como las cadenas no son iguales, obtenemos un valor de **Falso**.

```python
# Use Equality sign to compare the strings

"ACDC" == "The Bodyguard"
```

Si usamos el operador de desigualdad, la salida será **Verdadero** ya que las cadenas no son iguales.

```python
# Use Inequality sign to compare the strings

"ACDC" != "The Bodyguard"
```

### Ramificación

La ramificación nos permite ejecutar diferentes sentencias para distintas entradas. Es útil pensar en una sentencia if como una habitación cerrada: si la sentencia es verdadera, podemos entrar y el programa ejecutará algunas tareas predefinidas; pero si es falsa, el programa ignorará la tarea.

Por ejemplo, considere el rectángulo azul que representa un concierto de AC/DC. Si la persona es mayor de 18 años, puede entrar. Si tiene 18 años o menos, no puede.

Podemos usar las sentencias condicionales aprendidas anteriormente como las condiciones que deben verificarse en la sentencia if. La sintaxis es tan simple como <code> if <i>declaración condicional</i> :</code>, que contiene la palabra <code>if</code>, cualquier sentencia condicional y dos puntos al final. Inicie las tareas que deban ejecutarse bajo esta condición en una nueva línea con sangría. Las líneas de código después de los dos puntos y con sangría solo se ejecutarán cuando la sentencia if sea True. Las tareas finalizarán cuando la línea de código no contenga la sangría.

En el siguiente caso, el código <code>print(“you can enter”)</code> se ejecuta solo si la variable <code>age</code> es mayor que 18; es un caso True porque esta línea de código tiene la sangría. Sin embargo, la ejecución de <code>print(“move on”)</code> no se verá influenciada por la declaración if.

```python
# Ejemplo de sentencia if

edad = 19
#edad = 18

#Expresión que puede ser verdadera o falsa
if edad > 18:

#Dentro de una sangría, tenemos la expresión que se ejecuta si la condición es verdadera
    print("puedes entrar")

#Las sentencias posteriores a la sentencia if se ejecutarán independientemente de si la condición es verdadera o falsa
print("seguir adelante")
```

Intente descomentar la variable de edad.

Resulta útil usar el siguiente diagrama para ilustrar el proceso. A la izquierda, vemos qué sucede cuando la condición es <b>Verdadera</b>. La persona entra al concierto de AC/DC que representa el código en la sangría que se está ejecutando; luego, continúa. A la derecha, vemos qué sucede cuando la condición es <b>Falsa</b>; la persona no obtiene acceso y continúa. En este caso, el segmento de código en la sangría no se ejecuta, pero sí el resto de las instrucciones.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsIf.gif" width="650">

The <code>else</code> statement runs a block of code if none of the conditions are **True** before this <code>else</code> statement. Let's use the ACDC concert analogy again. If the user is 17 they cannot go to the ACDC concert,  but they can go to the Meatloaf concert.
The syntax of the <code>else</code> statement is similar as the syntax of the <code>if</code> statement, as <code>else :</code>. Notice that, there is no condition statement for <code>else</code>.
Try changing the values of <code>age</code> to see what happens:

```python
# Else statement example

age = 18
# age = 19

if age > 18:
    print("you can enter" )
else:
    print("go see Meat Loaf" )
    
print("move on")
```

El proceso se muestra a continuación, donde cada una de las posibilidades se ilustra a cada lado de la imagen. A la izquierda se muestra el caso en el que la edad es 17 años; establecemos la variable edad en 17, lo que corresponde a la persona que asiste al concierto de Meatloaf. La parte derecha muestra qué sucede cuando la persona es mayor de 18 años, en este caso 19, y se le permite el acceso al concierto.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsElse.gif" width="650">


La sentencia <code>elif</code>, abreviatura de else if, permite comprobar condiciones adicionales si las sentencias condicionales anteriores son <b>False</b>. Si la condición de la sentencia <code>elif</code> es <b>True</b>, se ejecutarán las expresiones alternativas. Consideremos el ejemplo del concierto: si una persona tiene 18 años, irá al concierto de Pink Floyd en lugar del de AC/DC o Meat-Loaf. Una persona de 18 años entra en el área y, como no tiene más de 18 años, no puede ver a AC/DC, pero como tiene 18 años, asiste a Pink Floyd. Después de ver a Pink Floyd, se marcha. La sintaxis de la sentencia <code>elif</code> es similar, ya que simplemente cambiamos el <code>if</code> en la sentencia <code>if</code> por <code>elif</code>.

```python
# Elif statment example

age = 18

if age > 18:
    print("you can enter" )
elif age == 18:
    print("go see Pink Floyd")
else:
    print("go see Meat Loaf" )
    
print("move on")
```

Las tres combinaciones se muestran en la figura a continuación. La región del extremo izquierdo muestra lo que sucede cuando el individuo es menor de 18 años. El componente central muestra cuándo el individuo tiene exactamente 18 años. El del extremo derecho muestra cuándo el individuo es mayor de 18 años.

Mira el siguiente código:

```python
# Condition statement example

album_year = 1983
album_year = 1970

if album_year > 1980:
    print("Album year is greater than 1980")
    
print('do something..')
```

Como antes, podemos añadir un bloque <code>else</code> al bloque <code>if</code>. El código del bloque <code>else</code> solo se ejecutará si el resultado es <b>Falso</b>.

Sintaxis:

if (condición):
\# hacer algo

### Logical operators

A veces quieres comprobar más de una condición al mismo tiempo. Por ejemplo, podrías querer verificar si una condición y otra condición son ambas **True**. Los operadores lógicos te permiten combinar o modificar condiciones.

<ul> <li><code>and</code></li> <li><code>or</code></li> <li><code>not</code></li> </ul>

Estos operadores se resumen para dos variables usando las siguientes tablas de verdad:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsTable.png" width="650">


La sentencia `and` solo es **True** cuando ambas condiciones son verdaderas.  
La sentencia `or` es **True** si una condición, o ambas, son **True**.  
La sentencia `not` devuelve el valor de verdad opuesto.

Veamos cómo determinar si un álbum fue lanzado después de 1979 (1979 no está incluido) y antes de 1990 (1990 no está incluido).  
Los períodos de tiempo entre 1980 y 1989 cumplen esta condición. Esto se demuestra en la figura de abajo.  

El color verde en las líneas **a** y **b** representa los períodos en los que la sentencia es **True**.  
El color verde en la línea **c** representa cuando ambas condiciones son **True**, lo cual corresponde a donde las regiones verdes se superponen.


<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsEgOne.png" width="650">

El bloque de código para realizar esta comprobación viene dado por:


```python
# Condition statement example

album_year = 2980

if(album_year > 1979) and (album_year < 1990):
    print ("Album year was in between 1980 and 1989")
    
print("")
print("Do Stuff..")
```

Para determinar si un álbum fue lanzado antes de 1980 (1979 y anteriores) o después de 1989 (1990 en adelante), se puede usar una sentencia `or`.  
Los períodos anteriores a 1980 (1979 y anteriores) o posteriores a 1989 (1990 en adelante) cumplen esta condición.  

Esto se muestra en la siguiente figura: el color verde en **a** y **b** representa los períodos en los que la sentencia es **True**.  
El color verde en **c** representa donde al menos una de las condiciones es **True**.

```python
# Condition statement example

album_year = 1990

if(album_year < 1980) or (album_year > 1989):
    print ("Album was not made in the 1980's")
else:
    print("The Album was made in the 1980's ")
```

La sentencia `not` comprueba si la condición es falsa:


```python
# Condition statement example

album_year = 1983

if not (album_year == 1984):
    print ("Album year is not 1984")
```
