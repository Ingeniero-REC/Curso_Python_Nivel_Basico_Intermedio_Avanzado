## Declaraciones de condición

### Operadores de comparación

Las **operaciones de comparación** comparan un valor u operando y, según una condición, producen un valor booleano (`True` o `False`). Al comparar dos valores, puedes utilizar los siguientes operadores:

<ul>
    <li>igual: <b>==</b></li>
    <li>diferente de: <b>!=</b></li>
    <li>mayor que: <b>></b></li>
    <li>menor que: <b>&lt;</b></li>
    <li>mayor o igual que: <b>>=</b></li>
    <li>menor o igual que: <b>&lt;=</b></li>
</ul>

Asignemos a <code>a</code> el valor de 5. Utiliza el operador de igualdad, representado por dos signos iguales <b>==</b>, para determinar si dos valores son iguales. En el siguiente caso, se compara la variable <code>a</code> con el valor 6.

```python
# Condición de Igualdad

a = 5
a == 6
```

El resultado es <b>False</b>, ya que 5 no es igual a 6.

Considera el siguiente operador de comparación: <code>i > 5</code>. Si el valor del operando de la izquierda, en este caso la variable <b>i</b>, es mayor que el valor del operando de la derecha, en este caso 5, entonces la expresión es <b>True</b>. De lo contrario, la expresión es <b>False</b>. Si <b>i</b> es igual a 6, como 6 es mayor que 5, el resultado será <b>True</b>.

```python
# Operador Mayor que

i = 6
i > 5
```

Establezca <code>i = 2</code>. La afirmación es falsa porque 2 no es mayor que 5:

```python
# Signo Mayor que

i = 2
i > 5
```

Mostremos algunos valores de <code>i</code> en la figura. Coloca en verde los valores mayores que 5 y en rojo el resto. La región verde representa donde la condición es **True**, mientras que la región roja representa donde la expresión es **False**. Si el valor de <code>i</code> es 2, obtenemos **False**, ya que 2 se encuentra en la región roja. De manera similar, si el valor de <code>i</code> es 6, obtenemos **True**, ya que la condición se encuentra en la región verde.

La prueba de desigualdad utiliza un signo de exclamación antes del signo igual. Si dos operandos no son iguales, entonces la condición será **True**. Por ejemplo, la siguiente condición producirá **True** siempre que el valor de <code>i</code> no sea igual a 6:

```python
# Operador de Desigualdad

i = 2
i != 6
```

Cuando <code>i</code> es igual a 6 la expresión de desigualdad produce <b>Falso</b>.

```python
# Operador de Desigualdad

i = 6
i != 6
```

Vea la recta numérica a continuación. Cuando la condición es **Verdadera**, los números correspondientes se marcan en verde, y si la condición es **Falsa**, el número correspondiente se marca en rojo. Si igualamos <code>i</code> a 2, el operador es verdadero, ya que 2 está en la región verde. Si igualamos <code>i</code> a 6, obtenemos un **Falso**, ya que la condición está en la región roja.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsIneq.gif" width="350">

```python
# Utiliza el operador de igualdad para comparar las cadenas de texto

"ACDC" == "The Bodyguard"
```

Si usamos el operador de desigualdad, la salida será **Verdadero** ya que las cadenas no son iguales.

```python
# Utiliza el operador de desigualdad para comparar las cadenas de texto

"ACDC" != "The Bodyguard"
```

### Ramificación

La ramificación nos permite ejecutar diferentes sentencias para distintas entradas. Es útil pensar en una sentencia if como una habitación cerrada: si la sentencia es verdadera, podemos entrar y el programa ejecutará algunas tareas predefinidas; pero si es falsa, el programa ignorará la tarea.

Por ejemplo, consideremos el rectángulo azul que representa un concierto de ACDC. Si la persona es mayor de 18 años, puede ingresar al concierto. Si tiene 18 años o menos, no puede ingresar al concierto.

Podemos utilizar las declaraciones de condición aprendidas anteriormente como las condiciones que deben verificarse en la **sentencia if**. La sintaxis es tan simple como <code> if <i>condition statement</i> :</code>, que contiene la palabra <code>if</code>, cualquier declaración de condición y dos puntos al final. Inicia las tareas que deben ejecutarse bajo esta condición en una nueva línea con una sangría. Las líneas de código después de los dos puntos y con sangría solo se ejecutarán cuando la **sentencia if** sea **True**. Las tareas finalizarán cuando la línea de código no contenga la sangría.

En el siguiente caso, el código <code>print(“you can enter”)</code> se ejecuta únicamente si la variable <code>age</code> es mayor que 18 y la condición es verdadera, porque esta línea de código tiene sangría. Sin embargo, la ejecución de <code>print(“move on”)</code> no se verá afectada por la sentencia if.

```python
# Ejemplo de sentencia if

edad = 19
# edad = 18

# Expresión que puede ser verdadera o falsa
if edad > 18:

    # Dentro de una sangría, tenemos la expresión que se ejecuta
    # si la condición es verdadera
    print("puedes entrar")

# Las sentencias posteriores a la sentencia if se ejecutarán
# independientemente de si la condición es verdadera o falsa
print("seguir adelante")
```

Intente descomentar la variable de edad.

Es útil utilizar el siguiente diagrama para ilustrar el proceso. En el lado izquierdo, vemos lo que sucede cuando la condición es <b>True</b>. La persona entra al concierto de ACDC, lo que representa que el código dentro de la sangría se ejecuta; después continúa adelante.

En el lado derecho, vemos lo que sucede cuando la condición es <b>False</b>; a la persona no se le permite el acceso y continúa adelante. En este caso, el segmento de código dentro de la sangría no se ejecuta, pero el resto de las instrucciones sí se ejecutan.

La sentencia <code>else</code> ejecuta un bloque de código si ninguna de las condiciones anteriores a esta sentencia <code>else</code> es **True**. Usemos nuevamente la analogía del concierto de ACDC. Si el usuario tiene 17 años, no puede ir al concierto de ACDC, pero sí puede ir al concierto de Meatloaf.

La sintaxis de la sentencia <code>else</code> es similar a la sintaxis de la sentencia <code>if</code>, como <code>else :</code>. Observa que no existe una declaración de condición para <code>else</code>.

Intenta cambiar los valores de <code>age</code> para ver qué sucede:

```python
# Ejemplo de sentencia else

edad = 18
# edad = 19

# Si la condición es verdadera, se ejecuta este bloque
if edad > 18:
    print("puedes entrar")

# Si la condición es falsa, se ejecuta este bloque
else:
    print("ve a ver a Meat Loaf")

# Esta línea siempre se ejecuta, independientemente del resultado del if-else
print("seguir adelante")
```

El proceso se muestra a continuación, donde cada una de las posibilidades se ilustra a cada lado de la imagen. A la izquierda se muestra el caso en el que la edad es 17 años; establecemos la variable edad en 17, lo que corresponde a la persona que asiste al concierto de Meatloaf. La parte derecha muestra qué sucede cuando la persona es mayor de 18 años, en este caso 19, y se le permite el acceso al concierto.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsElse.gif" width="450">


La sentencia <code>elif</code>, abreviatura de else if, permite comprobar condiciones adicionales si las sentencias condicionales anteriores son <b>False</b>. Si la condición de la sentencia <code>elif</code> es <b>True</b>, se ejecutarán las expresiones alternativas. Consideremos el ejemplo del concierto: si una persona tiene 18 años, irá al concierto de Pink Floyd en lugar del de AC/DC o Meat-Loaf. Una persona de 18 años entra en el área y, como no tiene más de 18 años, no puede ver a AC/DC, pero como tiene 18 años, asiste a Pink Floyd. Después de ver a Pink Floyd, se marcha. La sintaxis de la sentencia <code>elif</code> es similar, ya que simplemente cambiamos el <code>if</code> en la sentencia <code>if</code> por <code>elif</code>.

```python
# Ejemplo de sentencia elif

edad = 18

# Si la condición es verdadera, se ejecuta este bloque
if edad > 18:
    print("puedes entrar")

# Si la primera condición es falsa, se evalúa esta nueva condición
elif edad == 18:
    print("ve a ver a Pink Floyd")

# Si ninguna de las condiciones anteriores es verdadera, se ejecuta este bloque
else:
    print("ve a ver a Meat Loaf")

# Esta línea siempre se ejecuta al finalizar la estructura if-elif-else
print("seguir adelante")
```

Las tres combinaciones se muestran en la figura a continuación. La región del extremo izquierdo muestra lo que sucede cuando el individuo es menor de 18 años. El componente central muestra cuándo el individuo tiene exactamente 18 años. El del extremo derecho muestra cuándo el individuo es mayor de 18 años.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsElif.gif" width="450">

Mira el siguiente código:

```python
# Condition statement example

album_year = 1983
album_year = 1970

if album_year > 1980:
    print("Album year is greater than 1980")
    
print('do something..')
```

Como antes, podemos añadir un bloque <code>else</code> al bloque <code>if</code>. El código del bloque <code>else</code> solo se ejecutará si el resultado es <b>Falso</b>.

Sintaxis:

if (condición):
\# hacer algo

else:
\# hacer algo más

### Logical operators

A veces quieres comprobar más de una condición al mismo tiempo. Por ejemplo, podrías querer verificar si una condición y otra condición son ambas **True**. Los operadores lógicos te permiten combinar o modificar condiciones.

<ul> <li><code>and</code></li> <li><code>or</code></li> <li><code>not</code></li> </ul>

Estos operadores se resumen para dos variables usando las siguientes tablas de verdad:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsTable.png" width="650">


La sentencia `and` solo es **True** cuando ambas condiciones son verdaderas.  
La sentencia `or` es **True** si una condición, o ambas, son **True**.  
La sentencia `not` devuelve el valor de verdad opuesto.

Veamos cómo determinar si un álbum fue lanzado después de 1979 (1979 no está incluido) y antes de 1990 (1990 no está incluido).  
Los períodos de tiempo entre 1980 y 1989 cumplen esta condición. Esto se demuestra en la figura de abajo.  

El color verde en las líneas **a** y **b** representa los períodos en los que la sentencia es **True**.  
El color verde en la línea **c** representa cuando ambas condiciones son **True**, lo cual corresponde a donde las regiones verdes se superponen.


<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsEgOne.png" width="650">

El bloque de código para realizar esta comprobación viene dado por:


```python
# Condition statement example

album_year = 2980

if(album_year > 1979) and (album_year < 1990):
    print ("Album year was in between 1980 and 1989")
    
print("")
print("Do Stuff..")
```

Para determinar si un álbum fue lanzado antes de 1980 (1979 y anteriores) o después de 1989 (1990 en adelante), se puede usar una sentencia `or`.  
Los períodos anteriores a 1980 (1979 y anteriores) o posteriores a 1989 (1990 en adelante) cumplen esta condición.  

Esto se muestra en la siguiente figura: el color verde en **a** y **b** representa los períodos en los que la sentencia es **True**.  
El color verde en **c** representa donde al menos una de las condiciones es **True**.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%203/images/CondsEgTwo.png" width="650">

El bloque de código para realizar esta comprobación viene dado por:


```python
# Condition statement example

album_year = 1990

if(album_year < 1980) or (album_year > 1989):
    print ("Album was not made in the 1980's")
else:
    print("The Album was made in the 1980's ")
```

La sentencia `not` comprueba si la condición es falsa:


```python
# Condition statement example

album_year = 1983

if not (album_year == 1984):
    print ("Album year is not 1984")
```
