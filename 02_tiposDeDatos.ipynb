## Tipos de objetos en Python

Python es un lenguaje orientado a objetos. Existen muchos tipos de objetos en Python. Empecemos con los más comunes: cadenas, enteros y flotantes. Al escribir texto en Python, se utilizan cadenas de caracteres (cadenas, para abreviar). Los números más comunes, por otro lado, son los enteros (p. ej., -1, 0, 100) y los flotantes, que representan números reales (p. ej., 3,14, -42,0).

### Flotantes (Float)

Los números de coma flotante representan números reales; son un subconjunto de números enteros, pero también incluyen "números con decimales". Existen algunas limitaciones en cuanto a la representación de números reales por parte de máquinas, pero los números de coma flotante son una buena representación en la mayoría de los casos. Puede obtener más información sobre las características específicas de los números de coma flotante para su entorno de ejecución consultando el valor de <code>sys.float_info</code>. Esto también le indicará cuál es el número máximo y el mínimo que se puede representar con ellos.

<p>De nuevo, puede probar algunos ejemplos con la función <code>type()</code>:

````python
# Imprime el tipo de 1.0
# Ten en cuenta que 1 es un entero y 1.0 es un flotante
type(1.0)
````

````python
# Imprime el tipo de 0.5
type(0.5)
````

````python
# Imprimir el tipo de 0.56

type(0.56)
````

````python
# Configuración del sistema sobre el tipo de flotante
import sys
sys.float_info
````

### Conversión de un tipo de objeto a un tipo de objeto diferente

Puedes cambiar el tipo de un objeto en Python; esto se llama conversión de tipos. Por ejemplo, puedes convertir un entero en un flotante (p. ej., 2 en 2.0).

````python
# Verificar que sea un entero

type(2)
````

#### Conversión de enteros a flotantes
<p>Convirtamos el entero 2 a flotante:</p>

````python
# Convertir 2 en un valor flotante
float(2)
````

````python
# Convertir el entero 2 a un valor de punto flotante y comprobar su tipo
type(float(2))
````

Al convertir un entero a un número de punto flotante, no modificamos realmente el valor (es decir, la mantisa) del número. Sin embargo, si convertimos un número de punto flotante a un entero, podríamos perder información. Por ejemplo, si convertimos el número de punto flotante 1.1 a un entero, obtendremos 1 y perderemos la información decimal (es decir, 0.1).

````python
# Convertir 1.1 a entero resultará en pérdida de información
int(1.1)
````

````python
float(1)
````

### Conversión de cadenas a números enteros o flotantes

A veces, podemos tener una cadena que contiene un número. En ese caso, podemos convertir esa cadena que representa un número en un entero usando <code>int()</code>:</p>

````python
# Convertir una cadena en un entero
int('1')
````

Pero si intentas hacerlo con una cadena que no coincide exactamente con un número, obtendrás un error. Intenta lo siguiente:

````python
# Convierte una cadena en un entero con error
float("1 o 2 personas")
````

<p>También puedes convertir cadenas que contienen números de punto flotante en objetos <i>float</i>:</p>

````python
# Convierte la cadena "1.2" en un valor de punto flotante
float("1.2")
````

<hr/>
<div class="alert alert-success alertsuccess" style="margin-top: 20px">

### Tipo de datos booleano

<p><i>Booleano</i> es otro tipo importante en Python. Un objeto de tipo <i>Booleano</i> puede tomar uno de dos valores: <code>Verdadero</code> o <code>Falso</code>:</p>

````python
# Valor verdadero
True
````

Observe que el valor <code>True</code> tiene una "T" mayúscula. Lo mismo ocurre con <code>False</code> (es decir, debe usar la "F" mayúscula).

````python
# Valor falso
False
````

<p>Cuando le pides a Python que muestre el tipo de un objeto booleano, mostrará <code>bool</code> que significa <i>booleano</i>:</p>

````python
# Tipo de Verdadero
type(True)
````

````python
# Tipo de False
type(False)
````

Podemos convertir objetos booleanos a otros tipos de datos. Si convertimos un booleano con valor <code>True</code> a un entero o un número de punto flotante, obtendremos un uno. Si convertimos un booleano con valor <code>False</code> a un entero o un número de punto flotante, obtendremos un cero. De forma similar, si convertimos un 1 a un booleano, obtendremos un <code>True</code>. Y si convertimos un 0 a un booleano, obtendremos un <code>False</code>. Intentémoslo:

````python
# Convertir verdadero a entero
int(True)
````

````python
# Convertir 1 a booleano
bool(1)
````

````python
# Convertir falso a entero
int(False)
````

````python
# Convertir 0 a booleano
bool(0)
````

````python
# Convertir verdadero a flotante
float(True)
````
