<h2 id="tuple">Tuplas</h2>

En Python, existen diferentes tipos de datos: cadena, entero y flotante. Todos estos tipos de datos pueden incluirse en una tupla, como se indica a continuación:

<h2 id="dataset">Acerca del conjunto de datos</h2>

En Python, una tupla (tuple) es una colección ordenada e inmutable de elementos.

✔️ Características principales

Ordenada: los elementos mantienen su posición.

Inmutable: NO se pueden modificar después de crearla (no puedes añadir, borrar o cambiar elementos).

Permite elementos duplicados.

Puede contener distintos tipos de datos (int, str, listas, etc.).

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesType.png" width="750" align="center">


Ahora, vamos a crear nuestra primera tupla con cadena, número entero y punto flotante.

````python
# Crea tu primera tupla
tuple1 = ("disco",10, 1.2)
print(tuple1)
````

El tipo de variable es una **tupla**.

````python
# Imprime el tipo de tupla que creaste
type(tuple1)
````

✔️ ¿Por qué usar tuplas?

Son más rápidas que las listas.

Se usan para datos que no deben cambiar.

Útiles para retornar múltiples valores desde una función

# Indexación de una tupla

Se puede acceder a cada elemento de una tupla mediante un índice. La siguiente tabla representa la relación entre el índice y los elementos de la tupla. Cada elemento se puede obtener mediante el nombre de la tupla seguido de un corchete con el número de índice:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesIndex.gif" width="750" align="center">

Podemos imprimir cada valor en la tupla:

````python
# Imprimir la variable en cada índice
print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
````

Podemos imprimir el **tipo** de cada valor en la tupla:

````python
# Imprimir el tipo de cada dato de la tupla
print(type(tuple1[0]))
print(type(tuple1[1]))
print(type(tuple1[2]))
````

También podemos usar indexación negativa. Usamos la misma tabla anterior con los valores negativos correspondientes

<h3 id="sort">Ordenación</h3>

Considere la siguiente tupla:

````python
# Una tupla de muestra
Ratings = (0, 9, 6, 5, 10, 8, 9, 6, 2)
````

Podemos ordenar los valores de una tupla y guardarla en un nuevo objeto. El resultado es una nueva lista ordenada.

````python
# Ordenar la tupla
RatingsSorted = sorted(Ratings)
RatingsSorted
````

<h3 id="nest">Tupla anidada</h3>

Una tupla puede contener otra tupla, así como otros tipos de datos más complejos. Este proceso se denomina "anidación". Considere la siguiente tupla con varios elementos:

````python
# Crear una tupla de nido
NestedT =(1, 2, ("pop", "rock") ,(3,4),("disco",(1,2)))
````

Cada elemento de la tupla, incluidas otras tuplas, se puede obtener a través de un índice como se muestra en la figura:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesNestOne.png" width="750" align="center">

````python
#Imprimir elemento en cada índice
print("Element 0 of Tuple: ", NestedT[0])
print("Element 1 of Tuple: ", NestedT[1])
print("Element 2 of Tuple: ", NestedT[2])
print("Element 3 of Tuple: ", NestedT[3])
print("Element 4 of Tuple: ", NestedT[4])
````

Podemos utilizar el segundo índice para acceder a otras tuplas como se muestra en la figura:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesNestTwo.png" width="750" align="center">

Nosotros podemos acceder a las tuplas anidadas:

````python
# Imprimir elemento en cada índice, incluidos los índices anidados
print("Element 2, 0 of Tuple: ",   NestedT[2][0])
print("Element 2, 1 of Tuple: ",   NestedT[2][1])
print("Element 3, 0 of Tuple: ",   NestedT[3][0])
print("Element 3, 1 of Tuple: ",   NestedT[3][1])
print("Element 4, 0 of Tuple: ",   NestedT[4][0])
print("Element 4, 1 of Tuple: ",   NestedT[4][1])
````

Podemos acceder a las cadenas en las segundas tuplas anidadas usando un tercer índice:

````python
# Imprime el primer elemento en la segunda tupla anidada
NestedT[2][1][0]
````

````python
# Imprime el segundo elemento en la segunda tupla anidada
NestedT[2][1][1]
````

Podemos usar un árbol para visualizar el proceso. Cada nuevo índice corresponde a un nivel más profundo del árbol:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesNestThree.gif" width="750" align="center">

De manera similar, podemos acceder a elementos anidados más profundamente en el árbol con un tercer índice:

````python
# Imprime el primer elemento en la segunda tupla anidada
NestedT[4][1][0]
````

````python
# Imprime el segundo elemento en la segunda tupla anidada
NestedT[4][1][1]
````
