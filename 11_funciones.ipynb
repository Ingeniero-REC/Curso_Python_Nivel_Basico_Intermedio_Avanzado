<h2 id="func">Funciones</h2>

Una función es un bloque de código reutilizable que realiza las operaciones especificadas en la función. Las funciones permiten dividir las tareas en partes más pequeñas y reutilizar el código en diferentes programas.

Existen dos tipos de funciones:

* <b>Funciones predefinidas</b>
* <b>Funciones definidas por el usuario</b>

<h3 id="content">¿Qué es una Función?</h3>

Puedes definir funciones para proporcionar la funcionalidad requerida. Estas son algunas reglas básicas para definir una función en Python:

* Los bloques de funciones comienzan con <code>def</code>, seguido del <code>nombre</code> de la función y paréntesis <code>()</code>.
* Los parámetros de entrada o argumentos deben colocarse dentro de estos paréntesis.
* También puedes definir parámetros dentro de estos paréntesis.
* Cada función contiene un cuerpo que comienza con dos puntos (<code>:</code>) y está indentado.
* También puedes incluir documentación antes del cuerpo de la función.
* La sentencia <code>return</code> finaliza la ejecución de una función y, opcionalmente, devuelve un valor.

Un ejemplo de una función que suma un valor al parámetro <code>a</code>, imprime el resultado y lo devuelve como <code>b</code>:

````python
# Primer ejemplo de función: Sumar 1  con "a"  y almacenar el resultado en "b"

def add(a):
    """
    add 1 to a
    """
    b = a + 1
    print(a, "if you add one", b)
    return(b)
````

Podemos crear diferentes funciones. Por ejemplo, podemos crear una función que multiplique dos números. Los números estarán representados por las variables <code>a</code> y <code>b</code>:

<h3 id="var">Variables</h3>

La entrada de una función se denomina <b>parámetro formal</b>.

Una variable que se declara dentro de una función se denomina <b>variable local</b>. El parámetro solo existe dentro de la función (es decir, desde el punto donde la función comienza hasta donde termina).

Una variable que se declara fuera de la definición de una función es una <b>variable global</b>, y su valor es accesible y modificable en todo el programa. Hablaremos más sobre las variables globales al final del laboratorio.

````python
# Definición de la función

def square(a):
    
    # Variable local b
    b = 1
    c = a * a + b
    print(a, "si elevas al cuadrado + 1 es", c) 
    return(c)
````

## Uso de instrucciones if/else y bucles en funciones

La función <code>return()</code> es especialmente útil si tienes instrucciones IF dentro de la función, cuando deseas que la salida dependa de alguna condición:

````python
# Ejemplo de función

def type_of_album(album, year_released):
    
    print(album, year_released)
    if year_released > 1980:
        return "Moderno"
    else:
        return "Antiguo"
    
x = type_of_album("The BodyGuard", 1980)
print(x)
````

<h2 id="if">Comparación de cadenas de texto en funciones</h2>

Los operadores relacionales comparan los valores Unicode de los caracteres de las cadenas de texto desde el índice cero hasta el final de la cadena. Luego, devuelven un valor booleano según el operador utilizado.

````python
# Comparar dos cadenas directamente usando el operador in
# Añadir cadena de texto
string = "The BodyGuard is the best album"

# Definir una función
def check_string(text):
    
    # Usar la instrucción if else y el operador 'in' para comparar la cadena
    if text in string:
        return 'Cadena coincidente'
    else:
        return 'Cadena no coincidente'

check_string("The BodyGuard is the best")
````

Este programa utiliza una función definida por el usuario llamada <code>compareStrings()</code> para comparar dos cadenas de texto.

Esta función recibe ambas cadenas como argumentos y devuelve <code>1</code> si ambas cadenas son iguales utilizando el operador <code>==</code>.

````python
# Comparar dos cadenas usando el operador == y una función
def compareStrings(x, y):
    # Usar la instrucción if else para comparar x e y
    if x == y:
        return 1
    
# Declarar dos variables diferentes como string1 y string2 y pasar una cadena de texto en ellas
string1 = "The BodyGuard is the best album"
string2 = "The BodyGuard is the best album"

# Declarar una variable para almacenar el resultado después de comparar ambas cadenas
check = compareStrings(string1, string2)

# Usar la instrucción if else para comparar las cadenas
if check == 1:
    print("\nCadenas coincidentes")
else:
    print("\nCadenas no coincidentes")
````

**Contar la frecuencia de palabras que aparecen en una cadena de texto usando un diccionario.**

Encuentra el número de veces que aparece cualquier palabra en nuestra cadena de texto utilizando Python. Esto es lo que vamos a realizar en esta sección: contar la cantidad de palabras en una cadena dada e imprimir el resultado.

Supongamos que tenemos una "cadena" y una "palabra", y necesitamos encontrar la cantidad de veces que aparece esta palabra en nuestra cadena utilizando Python. Esto es lo que vamos a realizar en esta sección: contar el número de veces que aparece una palabra en una cadena dada e imprimirlo.

Lo primero que haremos será definir una función y luego crear una lista que estará vacía inicialmente.

A continuación, agregaremos código para convertir la cadena de texto en una lista. Las cadenas de Python tienen un método <code>split()</code>. Este método toma una cadena de texto y un separador, y devuelve una lista.

Ahora declararemos un diccionario vacío.

Después, agregaremos código utilizando un bucle <code>for</code> para recorrer las palabras. El valor contará la frecuencia de cada palabra en la cadena de texto y las almacenará en el diccionario.

Finalmente, imprimiremos el diccionario.

````python
# Programa en Python para contar palabras en una cadena usando un diccionario
def freq(string):
    
    # paso1: Se declara una variable de tipo lista y se inicializa como una lista vacía.
    words = []
    
    # paso2: Dividir la cadena de texto en una lista de palabras
    words = string.split() # o string.lower().split()
    
    # paso3: Declarar un diccionario
    Dict = {}
    
    # paso4: Usar un bucle for para recorrer las palabras y asignar valores al diccionario
    for key in words:
        Dict[key] = words.count(key)
        
    # paso5: Imprimir el diccionario
    print("La frecuencia de las palabras es:", Dict)
    
# paso6: Llamar a la función y pasar la cadena de texto como argumento
freq("Mary had a little lamb Little lamb, little lamb Mary had a little lamb.Its fleece was white as snow And everywhere that Mary went Mary went, Mary went \
Everywhere that Mary went The lamb was sure to go")
````

<h2 id="global">Variables globales</h2>

Hasta ahora, hemos estado creando variables dentro de las funciones, pero no hemos hablado sobre las variables que se encuentran fuera de una función. Estas se llaman **variables globales**. <br>
Veamos qué devuelve <code>printer1</code>:

````python
# Ejemplo de variable global

album = "The BodyGuard"

def printer1(album):
    internal_var1 = "Thriller"
    print(album, "es un álbum")
    
printer1(album)

# Intenta ejecutar el siguiente código
# printer1(internal_var1)
````

<b>Obtuvimos un NameError: <code>name 'internal_var' is not defined</code>. ¿Por qué?</b>

Esto se debe a que todas las variables que creamos dentro de una función son <b>variables locales</b>, lo que significa que la asignación de la variable no persiste fuera de la función.

Sin embargo, existe una forma de crear <b>variables globales</b> desde dentro de una función, como se muestra a continuación:

````python
album = "The BodyGuard"

def printer(album):
    global internal_var
    internal_var = "Thriller"
    print(album, "es un álbum")

printer(album)
printer(internal_var)
````

<h2 id="scope">Ámbito de una variable</h2>

El ámbito de una variable es la parte del programa donde esa variable es accesible. Las variables que se declaran fuera de todas las definiciones de funciones, como la variable <code>myFavouriteBand</code> en el código que se muestra aquí, son accesibles desde cualquier lugar del programa. Como resultado, se dice que estas variables tienen **ámbito global** y se conocen como **variables globales**. <code>myFavouriteBand</code> es una variable global, por lo que es accesible desde dentro de la función <code>getBandRating</code>, y podemos utilizarla para determinar la calificación de una banda. También podemos usarla fuera de la función, por ejemplo, al pasarla a la función <code>print</code> para mostrar su valor:

````python
# Ejemplo de variable global

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("La calificación de AC/DC es:", getBandRating("AC/DC"))
print("La calificación de Deep Purple es:", getBandRating("Deep Purple"))
print("Mi banda favorita es:", myFavouriteBand)
````

Observa esta versión modificada de nuestro código. Ahora la variable <code>myFavouriteBand</code> está definida dentro de la función <code>getBandRating</code>. Una variable que se define dentro de una función se conoce como una **variable local** de esa función. Esto significa que solo es accesible desde el interior de la función en la que fue definida. Nuestra función <code>getBandRating</code> seguirá funcionando, porque <code>myFavouriteBand</code> todavía está definida dentro de la función. Sin embargo, ya no podemos imprimir <code>myFavouriteBand</code> fuera de nuestra función, porque es una variable local de la función <code>getBandRating</code>; solo está definida dentro de dicha función:

````python
# Eliminar la variable "myFavouriteBand" del ejemplo anterior para demostrar el uso de una variable local

del myFavouriteBand

# Ejemplo de variable local

def getBandRating(bandname):
    myFavouriteBand = "AC/DC"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("La calificación de AC/DC es:", getBandRating("AC/DC"))
print("La calificación de Deep Purple es:", getBandRating("Deep Purple"))
print("Mi banda favorita es:", myFavouriteBand)
````

Finalmente, observa este ejemplo. Ahora tenemos dos definiciones de la variable <code>myFavouriteBand</code>. La primera tiene **ámbito global**, y la segunda es una **variable local** dentro de la función <code>getBandRating</code>. Dentro de la función <code>getBandRating</code>, la variable local tiene prioridad. **Deep Purple** recibirá una calificación de <code>10.0</code> cuando se pase a la función <code>getBandRating</code>. Sin embargo, fuera de la función <code>getBandRating</code>, la variable local de <code>getBandRating</code> no está definida, por lo que la variable <code>myFavouriteBand</code> que imprimimos es la variable global, cuyo valor es **AC/DC**:

````python
# Ejemplo de una variable global y una variable local con el mismo nombre

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    myFavouriteBand = "Deep Purple"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("La calificación de AC/DC es:", getBandRating("AC/DC"))
print("La calificación de Deep Purple es:", getBandRating("Deep Purple"))
print("Mi banda favorita es:", myFavouriteBand)
````

## Colecciones y funciones

Cuando se desconoce el número de argumentos que recibirá una función, todos ellos pueden agruparse en una tupla, como se muestra a continuación:

````python
def printAll(*args):  # Todos los argumentos se "agrupan" en args, que puede tratarse como una tupla
    print("Número de argumentos:", len(args))
    for argument in args:
        print(argument)

# printAll con 3 argumentos
printAll('Horsefeather', 'Adonis', 'Bone')

# printAll con 4 argumentos
printAll('Sidecar', 'Long Island', 'Mudslide', 'Carriage')
````

De manera similar, los argumentos también pueden agruparse en un diccionario, como se muestra a continuación:

````python
def printDictionary(**args):
    for key in args:
        print(key + " : " + args[key])

printDictionary(Country='Canada',Province='Ontario',City='Toronto')
````

Las funciones pueden ser increíblemente potentes y versátiles. Pueden aceptar (y devolver) tipos de datos, objetos e incluso otras funciones como argumentos. Considera el siguiente ejemplo:

````python
def addItems(list):
    list.append("Three")
    list.append("Four")

myList = ["One","Two"]

addItems(myList)

myList
    
````

Observa que los cambios realizados en la lista no están limitados al ámbito de la función. Esto ocurre porque lo que se pasa a la función es la **referencia** de la lista; cualquier cambio realizado afecta a la instancia original de la lista. Por lo tanto, se debe tener cuidado al pasar objetos mutables a las funciones.

```python
# Programa en Python para contar palabras en una cadena usando un diccionario
def freq(string, passedkey):

    # paso1: Se declara una variable de tipo lista y se inicializa como una lista vacía.
    words = []

    # paso2: Dividir la cadena de texto en una lista de palabras
    words = string.split() # o string.lower().split()

    # paso3: Declarar un diccionario
    Dict = {}

    # paso4: Usar un bucle for para recorrer las palabras y asignar valores al diccionario
    for key in words:
        if(key == passedkey):
            Dict[key] = words.count(key)

    # paso5: Imprimir el diccionario
    print("Conteo total:", Dict)

# paso6: Llamar a la función y pasar la cadena de texto como argumento
freq("Mary had a little lamb Little lamb, little lamb Mary had a little lamb.Its fleece was white as snow And everywhere that Mary went Mary went, Mary went \
Everywhere that Mary went The lamb was sure to go", "little")
