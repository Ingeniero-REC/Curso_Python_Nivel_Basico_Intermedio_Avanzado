## Sets

### Set Content

Un conjunto es una colección única de objetos en Python. Puedes representar un conjunto con un par de llaves <b>{}</b>. Python eliminará automáticamente los elementos duplicados:

````python
# Creacion de un set(conjunto)
set1={"pop","rock","soul","hard rock","rock","R&B","rock","disco"}
set1
````

El proceso de mapeo se ilustra en la figura:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsUnique.png" width="500">

También puedes crear un conjunto a partir de una lista de la siguiente manera:

````python
# Convertir lista a conjunto

album_list = [ "Michael Jackson", "Thriller", 1982, "00:42:19", \
              "Pop, Rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
album_set = set(album_list)             
album_set
````

Ahora vamos a crear un conjunto de géneros:

````python
# Convertir lista a conjunto
music_genres = set(["pop", "pop", "rock", "folk rock", "hard rock", "soul", \
                    "progressive rock", "soft rock", "R&B", "disco"])
music_genres
````

### Set Operations 


Repasemos las operaciones con conjuntos, ya que pueden utilizarse para modificar el conjunto. Consideremos el conjunto <b>A</b>:

````python
#Set de muestra
A = set(["Thriller", "Back in Black", "AC/DC"])
A
````

Podemos agregar un elemento a un conjunto usando el método <code>add()</code>:

````python
#Aderir un elemento a un set
A.add("NSYNC")
A
````

Si agregamos el mismo elemento dos veces, no ocurrirá nada ya que no puede haber duplicados en un conjunto:

````python
#Aderir por segunda vez un elemento a un set
A.add("NSYNC")
A
````

Podemos eliminar un elemento de un conjunto usando el método <code>remove</code>:

````python
# Eliminar el elemento del conjunto
A.remove("NSYNC")
A
````

Podemos verificar si un elemento está en el conjunto usando el comando <code>in</code>:

````python
#Verificar si el elemento esta en el set
"AC/DC" in A 
````

### Sets Logic Operations


Recuerda que con los conjuntos puedes comprobar la diferencia entre conjuntos, así como la diferencia simétrica, la intersección y la unión:

Consideremos los dos SETS siguientes:

````python
# Muestra de Sets

album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
````

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsSamples.png" width="500">

````python
# Imprimir dos conjuntos

album_set1, album_set2
````

Como ambos sets contienen <b>AC/DC</b> y <b>Back in Black</b>, representamos estos elementos comunes con la intersección de dos círculos.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsLogic.png" width="450">

Puedes encontrar la intersección de dos conjuntos de la siguiente manera usando <code>&</code>:

````python
# Encontrar la intersección

intersection = album_set1 & album_set2
intersection
````

Puedes encontrar todos los elementos que solo están contenidos en <code>album_set1</code> usando el método <code>difference</code>:

````python
# Encontrar la diferencia: en set1 pero no en set2

album_set1.difference(album_set2) 
````

Solo debe considerar los elementos en <code>album_set1</code>; todos los elementos en <code>album_set2</code>, incluida la intersección, no están incluidos.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsLeft.png" width="450">

Los elementos en <code>album_set2</code> pero no en <code>album_set1</code> se obtienen mediante:

````python
album_set2.difference(album_set1)  
````

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsRight.png" width="450">

También puedes encontrar la intersección de <code>album_list1</code> y <code>album_list2</code> usando el método <code>intersection</code>:

````python
# Usa el método intersection para encontrar la intersección de album_list1 y album_list2

album_set1.intersection(album_set2)
````

Esto corresponde a la intersección de los dos círculos:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsIntersect.png" width="450">

La unión corresponde a todos los elementos de ambos conjuntos, lo que se representa coloreando ambos círculos:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsUnion.png" width="450">

La unión viene dada por:

````python
# Encontrar la unión de dos conjuntos

album_set1.union(album_set2)
````

Y puedes comprobar si un conjunto es un superconjunto o subconjunto de otro conjunto, respectivamente, de la siguiente manera:

````python
# Verificar si es superconjunto

set(album_set1).issuperset(album_set2)   
````

Aquí hay un ejemplo donde <code>issubset()</code> y <code>issuperset()</code> devuelven verdadero:

````python
# CVerificar si es subset

set({"Back in Black", "AC/DC"}).issubset(album_set1) 
````

````python
# Verificar si es superset

album_set1.issuperset({"Back in Black", "AC/DC"})   
````
