## Sets

### Set Content

Un conjunto es una colección única de objetos en Python. Se puede representar con llaves <b>{}</b>. Python eliminará automáticamente los elementos duplicados:

````python
# Creacion de un set(conjunto)
set1={"pop","rock","soul","hard rock","rock","R&B","rock","disco"}
set1
````

El proceso de mapeo se ilustra en la figura:

Ahora vamos a crear un conjunto de géneros:

````python
generos_musicales=set(["pop", "pop", "rock", "folk rock", "hard rock", "soul",
                    "progressive rock", "soft rock", "R&B", "disco"])
generos_musicales
````

### Set Operations 


Repasemos las operaciones con conjuntos, ya que pueden utilizarse para modificar el conjunto. Consideremos el conjunto <b>A</b>:

````python
#Set de muestra
A = set(["Thriller", "Back in Black", "AC/DC"])
A
````

Podemos agregar un elemento a un conjunto usando el método <code>add()</code>:

````python
#Aderir un elemento a un set
A.add("NSYNC")
A
````

Si agregamos el mismo elemento dos veces, no ocurrirá nada ya que no puede haber duplicados en un conjunto:

````python
#Aderir por segunda vez un elemento a un set
A.add("NSYNC")
A
````

Podemos verificar si un elemento está en el SET usando el comando in:

````python
#Verificar si el elemento esta en el set
"AC/DC" in A 
````

### Sets Logic Operations


Recuerda que con los conjuntos puedes comprobar la diferencia entre conjuntos, así como la diferencia simétrica, la intersección y la unión:

Consideremos los dos SETS siguientes:

````python
# Muestra de Sets

album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
````

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsSamples.png" width="650">

````python
# Print two sets

album_set1, album_set2
````

Como ambos sets contienen <b>AC/DC</b> y <b>Back in Black</b>, representamos estos elementos comunes con la intersección de dos círculos.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsLogic.png" width="650">

Puedes encontrar la intersección de dos conjuntos de la siguiente manera usando <code>&</code>:

````python
# Find the intersections(encontrar interseccion)

intersection = album_set1 & album_set2
intersection
````

Puedes encontrar todos los elementos que solo están contenidos en <code>album_set1</code> usando el método <code>difference</code>:

````python
# Find the difference in set1 but not set2

album_set1.difference(album_set2) 
````

Solo debe considerar los elementos en <code>album_set1</code>; todos los elementos en <code>album_set2</code>, incluida la intersección, no están incluidos.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsLeft.png" width="650">

Los elementos en album_set2 pero no en album_set1 se dan por:

````python
album_set2.difference(album_set1)  
````

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsRight.png" width="650">

También puedes encontrar la intersección de album_list1 y album_list2, utilizando el método de intersección:

````python
# Use intersection method to find the intersection of album_list1 and album_list2

album_set1.intersection(album_set2)   
````

Esto corresponde a la intersección de los dos círculos:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsIntersect.png" width="650">

La unión corresponde a todos los elementos de ambos conjuntos, lo que se representa coloreando ambos círculos:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsUnion.png" width="650">

La unión viene dada por:

````python
# Find the union of two sets

album_set1.union(album_set2)
````

Y puedes comprobar si un conjunto es un superconjunto o subconjunto de otro conjunto, respectivamente, de la siguiente manera:

````python
# Check if superset

set(album_set1).issuperset(album_set2)   
````

Aquí hay un ejemplo donde <code>issubset()</code> y <code>issuperset()</code> devuelven verdadero:

````python
# Check if subset

set({"Back in Black", "AC/DC"}).issubset(album_set1) 
````

````python
# Check if superset

album_set1.issuperset({"Back in Black", "AC/DC"})   
````
