## ¿Qué son los diccionarios?

Un diccionario consta de claves y valores. Resulta útil comparar un diccionario con una lista. En lugar de estar indexados numéricamente como una lista, los diccionarios tienen claves. Estas claves son las que se utilizan para acceder a los valores dentro de un diccionario.

El mejor ejemplo de un diccionario es acceder a los datos de una persona mediante el **número de la seguridad social**.
En este caso, el número de la seguridad social, que es un número único, será la **clave** y los datos de las personas serán los **valores** asociados a él.

### Crea un diccionario y accede a los elementos

Ejemplo de un diccionario <code>Dict</code>:
Aquí creamos un diccionario llamado **Dict** con los siguientes detalles:

* Las claves son **clave1, clave2, clave3, clave4, clave5, (0,1)**.
* Los valores son {1,2,[3,3,3],(4,4,4),5,6}, correspondientes a las claves.

````python
# Crea el diccionario
Dict = {"key1": 1, "key2": "2", "key3": [3, 3, 3], "key4": (4, 4, 4), ('key5'): 5, (0, 1): 6}
Dict
````

Las claves pueden ser cadenas:

````python
#Acceso al valor mediante la clave
Dict["key1"]
````

Las claves también pueden ser cualquier objeto inmutable, como una tupla:

````python
#Acceso al valor mediante la clave
Dict[(0, 1)]
````

Cada clave se separa de su valor con dos puntos "<code>:</code>". Los elementos se separan con comas y el diccionario completo se escribe entre llaves. Un diccionario vacío sin elementos se escribe con solo dos llaves, como este "<code>{}</code>".

````python
# Crear un diccionario de muestra

release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
````

En resumen, al igual que una lista, un diccionario contiene una secuencia de elementos. Cada elemento se representa mediante una clave y su valor correspondiente. Los diccionarios se crean con dos llaves que contienen claves y valores separados por dos puntos. Cada clave solo puede tener un valor; sin embargo, varias claves pueden contener el mismo valor. Las claves solo pueden ser cadenas, números o tuplas, pero los valores pueden ser de cualquier tipo de dato.

Resulta útil visualizar el diccionario como una tabla, como en la siguiente imagen. La primera columna representa las claves y la segunda, los valores.

## Claves

Puede recuperar los valores basándose en los nombres:

````python
# Obtener valor por claves
release_year_dict['Thriller'] 
````

Del mismo modo para <b>The Bodyguard</b>

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/DictsKeyTwo.png" width="500">


Ahora recuperemos las claves del diccionario usando el método <code>keys()</code>:

````python
# Obtener todas las claves del diccionario
release_year_dict.keys() 
````

Puede recuperar los valores utilizando el método <code>values()</code>:

````python
# Obtener todos los valores del diccionario
release_year_dict.values() 
````

Podemos añadir una entrada:

````python
# Añadir valor con clave al diccionario
release_year_dict['Graduation'] = '2007'
release_year_dict
````

Nosotros podemos borrar una entrada:

````python
# Eliminar entradas por clave
del(release_year_dict['Thriller'])
del(release_year_dict['Graduation'])
release_year_dict
````

Podemos verificar si un elemento está en el diccionario:

````python
# Verificar que la clave esté en el diccionario
'The Bodyguard' in release_year_dict
````
