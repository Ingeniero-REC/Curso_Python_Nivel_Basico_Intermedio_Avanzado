### Escribir comentarios en Python

<p>
Además de escribir código, ten en cuenta que siempre es buena idea añadir comentarios. Esto ayudará a otros a comprender lo que intentabas lograr (la razón por la que escribiste un fragmento de código). Esto no solo ayuda a que otras personas entiendan tu código, sino que también puede servirte de recordatorio cuando lo revises semanas o meses después.</p>

<p>
Para escribir comentarios en Python, usa el símbolo <code>#</code> antes de escribirlos. Al ejecutar tu código, Python ignorará todo lo que esté después del <code>#</code> en una línea determinada.
</p>

````python
# Práctica sobre cómo escribir comentarios

print('¡Hola, Python!')  # Esta línea imprime una cadena de texto
# print("Hola")
````

<p>
Después de ejecutar la celda anterior, deberías notar que <code>Esta línea imprime una cadena de texto</code> no apareció en la salida, porque era un comentario (y, por lo tanto, Python lo ignoró).
</p>
<p>
La segunda línea tampoco se ejecutó porque <code>print('Hola')</code> también estaba precedido por el signo de número (<code>#</code>). Dado que no se trata de un comentario explicativo del programador, sino de una línea de código real, podríamos decir que el programador comentó esa segunda línea de código.
</p>

### Errores en Python

Todos cometemos errores. Para muchos tipos de errores, Python te indicará que has cometido uno mediante un mensaje de error. Es importante leer los mensajes de error con atención para comprender realmente dónde cometiste un error y cómo puedes corregirlo.

Por ejemplo, si escribes <code>print</code> como <code>frint</code>, Python mostrará un mensaje de error. Inténtalo:

# Imprimir cadena como mensaje de error

````python
# Imprimir cadena como mensaje de error
frint("Richard")
````

El mensaje de error indica:

<ol>
<li>dónde se produjo el error (más útil en celdas grandes de cuadernos o scripts) y
<li>qué tipo de error fue (NameError)

</ol>
<p>Aquí, Python intentó ejecutar la función <code>frint</code>, pero no pudo determinar qué es, ya que no es una función integrada y tampoco la hemos definido previamente.

````python
# Intenta ver el mensaje de error integrado
print("Richard")
````

### ¿Python sabe acerca de su error antes de ejecutar su código?

Python es lo que se denomina un lenguaje interpretado. Los lenguajes compilados examinan el programa completo en tiempo de compilación y pueden advertir sobre toda una clase de errores antes de su ejecución. En cambio, Python interpreta el script línea por línea a medida que lo ejecuta. Python detendrá la ejecución del programa completo si encuentra un error (a menos que el programador lo espere y lo gestione, un tema más avanzado que abordaremos más adelante en este curso).

Intente ejecutar el código en la celda a continuación y vea qué sucede:

````python
# Imprime la cadena y el error para ver el orden de ejecución

print("Esto se imprimirá")
frint("Esto causará un error")
print("Esto NO se imprimirá")
````
