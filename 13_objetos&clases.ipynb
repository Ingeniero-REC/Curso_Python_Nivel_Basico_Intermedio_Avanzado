### Métodos

Los métodos te proporcionan una forma de cambiar o interactuar con el objeto; son funciones que interactúan con los objetos. Por ejemplo, supongamos que queremos aumentar el radio de un círculo en una cantidad específica. Podemos crear un método llamado **add_radius(r)** que incrementa el radio en **r**.

Esto se muestra en la figura 3, donde después de aplicar el método al "objeto círculo naranja", el radio del objeto aumenta en consecuencia.

La notación con “punto” significa aplicar el método al objeto, lo que básicamente consiste en aplicar una función a la información contenida en el objeto.

Ahora vamos a crear una clase Circle, pero primero vamos a importar una biblioteca para dibujar los objetos:

````python
# Importar la biblioteca

import matplotlib.pyplot as plt
%matplotlib inline
````

El primer paso para crear tu propia clase es utilizar la palabra clave <code>class</code>, seguida del nombre de la clase, como se muestra en la Figura 4. En este curso, la clase padre siempre será <code>object</code>:

El siguiente paso es un método especial llamado constructor <code>\__init\_\_</code>, que se utiliza para inicializar el objeto. Las entradas son atributos de datos.

El término <code>self</code> contiene todos los atributos del conjunto. Por ejemplo, <code>self.color</code> proporciona el valor del atributo color y <code>self.radius</code> proporciona el radio del objeto.

También tenemos el método <code>add_radius()</code> con el parámetro <code>r</code>; este método agrega el valor de <code>r</code> al atributo radius. Para acceder al radio utilizamos la sintaxis <code>self.radius</code>.

La sintaxis etiquetada se resume en la Figura 5:

<i>Figura 4: Creando una clase Circle.</i>

<i>Figura 5: Sintaxis etiquetada del objeto círculo.</i>

El objeto real se muestra a continuación. Incluimos el método <code>drawCircle</code> para mostrar la imagen de un círculo. Establecemos el radio predeterminado en 3 y el color predeterminado en azul:

````python
# Create a class Circle

class Circle(object):
    
    # Constructor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius
        self.color = color 
    
    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)
    
    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()  
````

````python
# Crear un círculo rojo con radio 5
circle2 = Circle(radius=5, color='red')
# Dibujar el círculo
circle2.drawCircle()
````

## La clase Rectangle

Vamos a crear una clase rectangle con los atributos height, width y color. Solo agregaremos el método para dibujar el objeto rectangle:

````python
# Crear una nueva clase Rectangle para crear un objeto rectangle

class Rectangle(object):
    
    # Constructor
    def __init__(self, width=2, height=3, color='r'):
        self.height = height 
        self.width = width
        self.color = color
    
    # Método
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height, fc=self.color))
        plt.axis('scaled')
        plt.show()
````

Vamos a crear el objeto <code>SkinnyBlueRectangle</code> de tipo Rectangle. Su ancho será 2, su altura será 3 y el color será azul:

````python
# Crear un nuevo objeto rectangle

SkinnyBlueRectangle = Rectangle(2, 3, 'blue')
````
