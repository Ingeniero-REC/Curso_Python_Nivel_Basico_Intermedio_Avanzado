## Listas

## Indexacion


Vamos a ver las listas en Python. Una lista es una colección ordenada de diferentes objetos, como enteros, cadenas de texto e incluso otras listas. La dirección de cada elemento dentro de una lista se llama <b>índice</b>. Un índice se utiliza para acceder y hacer referencia a los elementos dentro de una lista.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/ListsIndex.png" width="500">


Para crear una lista, escribe la lista entre corchetes <b>\[ ]</b>, con el contenido entre paréntesis y separado por comas. ¡Probémoslo!

````python
# Creemos una lista
L = ["The Bodyguard", 7.0, 1992 ]
L
````

Podemos utilizar indexación negativa y regular con una lista:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/kkMiHZTtmXzY4BWInK5Azg/Slide1.PNG" width="500">

````python
# Imprimir los elementos en cada índice
print('El mismo elemento usando indexacion positiva y negativa:\n Positivo:', L[0],'\n Negativo:', L[-3])
print('El mismo elemento usando indexacion positiva y negativa:\n Positivo:', L[1],'\n Negativo:', L[-2])
print('El mismo elemento usando indexacion positiva y negativa:\n Positivo:', L[2],'\n Negativo:', L[-1])
````

## Contenido de la lista

Las listas pueden contener cadenas, números de punto flotante y enteros. Podemos anidar otras listas, así como tuplas y otras estructuras de datos. Se aplican las mismas convenciones de indexación para la anidación:

````python
# Lista de ejemplos
["The Bodyguard", 7.0, 1992, [1, 2], ("A", 1)]
````

## Operaciones con listas

También podemos segmentar listas. Por ejemplo, si queremos los dos últimos elementos, usamos el siguiente comando:

````python
# Lista de muestra
L = ["The Bodyguard", 7.0,1992,"BG",1]
L
````

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/KyRT0sxnEHkVDzA5XS6h0w/Slide2.PNG" width="500">

````python
# Segmentación de listas (Slicing)
L[3:5]
````

Podemos usar el método <code>extend</code> para agregar nuevos elementos a la lista:

````python
# Utilice extender para agregar elementos a la lista
L = [ "The Bodyguard", 7.0]
L.extend(['pop', 10])
L
````

Otro método similar es <code>append</code>. Si aplicamos <code>append</code> en lugar de <code>extend</code>, añadimos una lista:

````python
# Utilice append para agregar elementos a la lista
L = [ "The Bodyguard", 7.0]
L.append(['pop', 10])
L
````

Cada vez que aplicamos un método, la lista cambia. Si aplicamos <code>extend</code>, añadimos dos elementos nuevos a la lista. La lista <code>L</code> se modifica añadiendo dos elementos nuevos:

````python
# Utilice extender para agregar elementos a la lista
L = [ "The Bodyguard", 7.0]
L.extend(['pop', 10])
L
````

Si agregamos la lista <code>\['a','b']</code> tenemos un nuevo elemento que consiste en una lista anidada:

````python
# Utilice append para agregar elementos a la lista
L.append(['a','b'])
L
````

Como las listas son mutables, podemos modificarlas. Por ejemplo, podemos cambiar el primer elemento de la siguiente manera:

````python
# Cambiar el elemento en función del índice
A = ["disco", 10, 1.2]
print('Antes del cambio:', A)
A[0] = 'hard rock'
print('Despues del cambio:', A)
````

También podemos eliminar un elemento de una lista usando el comando <code>del</code>:

````python
# Eliminar el elemento según el índice
print('Antes del cambio:', A)
del(A[0])
print('Despues del cambio:', A)
````

Podemos convertir una cadena en una lista usando <code>split</code>. Por ejemplo, el método <code>split</code> traduce cada grupo de caracteres separados por un espacio en un elemento de una lista:

````python
# Divide la cadena, el valor predeterminado es por espacio
'hard rock'.split()
````

Podemos usar la función split para separar cadenas por un carácter específico, llamado **delimitador**. Pasamos el carácter que queremos dividir al argumento, que en este caso es una coma. El resultado es una lista, y cada elemento corresponde a un conjunto de caracteres separados por una coma:

````python
# Dividir la cadena por coma
'A,B,C,D'.split(',')
````

## Copiar y clonar lista

Cuando establecemos una variable <b>B</b> igual a <b>A</b>, tanto <b>A</b> como <b>B</b> hacen referencia a la misma lista en la memoria:

````python
# Copiar (copiar por referencia) la lista A
A = ["hard rock", 10, 1.2]
B = A
print('A:', A)
print('B:', B)
````

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/ListsRef.png" width="500" align="center">


Inicialmente, el valor del primer elemento de <b>B</b> se establece como "hard rock". Si cambiamos el primer elemento de <b>A</b> a <b>"banana"</b>, obtenemos un efecto secundario inesperado. Como <b>A</b> y <b>B</b> hacen referencia a la misma lista, si cambiamos la lista <b>A</b>, la lista <b>B</b> también cambia. Si comprobamos el primer elemento de <b>B</b>, obtenemos "banana" en lugar de "hard rock":

````python
# Examinar la copia por referencia
print('B[0]:', B[0])
A[0] = "banana"
print('B[0]:', B[0])

print(A)
print(B)
````

Esto se demuestra en la siguiente figura:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/ListsRefGif.gif" width="500">

Puede clonar la lista **A** utilizando la siguiente sintaxis:

````python
# Clonar (clonar por valor) la lista A
B = A[:]
B
````

La variable **B** hace referencia a una nueva copia o clon de la lista original. Esto se muestra en la siguiente figura:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/ListsVal.gif" width="500">

Ahora bien, si cambias <b>A</b>, <b>B</b> no cambiará:

````python
print('B[0]:', B[0])
A[0] = "hard rock"
print('B[0]:', B[0])

print(A)
print(B)
 
````
