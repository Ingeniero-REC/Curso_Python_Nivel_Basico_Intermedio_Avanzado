### Expresiones

Las expresiones en Python pueden incluir operaciones entre tipos compatibles (p. ej., enteros y flotantes). Por ejemplo, operaciones aritméticas básicas como sumar varios números:

````python
# Expresión de operación de suma
43 + 60 + 16 + 41
````

Podemos realizar operaciones de resta usando el operador menos. En este caso, el resultado es un número negativo:

````python
# Expresión de operación de resta
50 - 60
````

Podemos hacer la multiplicación usando un asterisco(*):

````python
# Expresión de operación de multiplicación

5 * 5 
````

También podemos realizar la división con la barra diagonal (/):

````python
# Expresión de operación de división exacta
25 / 5 
````

````python
# Expresión de operación de división con decimales
25 / 6
````

<p>Como se ve en el cuestionario anterior, podemos usar la doble barra (//) para la división de números enteros, donde el resultado se redondea hacia abajo al número entero más cercano:

````python
# Expresión de operación de división de enteros
25 // 5
````

````python
# Expresión de operación de división de enteros
25 // 6
````

Escribamos una expresión que calcule cuántas horas hay en 160 minutos:

````python
# Escribe tu código a continuación. No olvides presionar Mayús+Intro para ejecutar la celda
160 // 60
````

Python sigue las convenciones matemáticas aceptadas al evaluar expresiones matemáticas. En el siguiente ejemplo, Python suma 30 al resultado de la multiplicación (es decir, 120).

````python
# Expresión matemática

30 + 2 * 60
````

Y al igual que en las matemáticas, las expresiones entre paréntesis tienen prioridad. Así que la siguiente multiplica 32 por 60.

````python
# Expresión matemática

(30 + 2) * 60 
````

### Variables


Al igual que con la mayoría de los lenguajes de programación, podemos almacenar valores en variables para usarlos posteriormente. Por ejemplo:

````python
# Almacenar valor en variable

x = 43 + 60 + 16 + 41
````

<p>Para ver el valor de <code>x</code> en un Notebook, simplemente podemos colocarlo en la última línea de una celda:</p>

````python
# Imprime el valor en la variable
x
````

<p>También podemos realizar operaciones en <code>x</code> y guardar el resultado en una nueva variable:</p>

````python
# Usar otra variable para almacenar el resultado de la operación entre la variable y el valor

y = x / 60
y
````

<p>Si guardamos un valor en una variable existente, el nuevo valor sobrescribirá el valor anterior:</p>

````python
# Sobrescribir la variable con el nuevo valor

x = x / 60
x = 5
print(x)
````

<p>Es una buena práctica utilizar nombres de variables significativos, para que usted y otros puedan leer el código y comprenderlo más fácilmente:</p>

````python
# Nombra las variables de forma significativa

total_min = 43 + 42 + 57 # Duración total de los álbumes en minutos
total_min
````

````python
# Nombra las variables de forma significativa

total_horas = total_min / 60 # Duración total de los álbumes en horas
total_horas
````

En las celdas anteriores, sumamos la duración de tres álbumes en minutos y la almacenamos en <code>total_min</code>. Luego, la dividimos entre 60 para calcular la duración total <code>total_hours</code> en horas. También puedes calcularlo todo a la vez en una sola expresión, siempre que uses paréntesis para sumar la duración de los álbumes antes de dividir, como se muestra a continuación.

````python
# Expresión ComplicateD

total_horas = (43 + 42 + 57) / 60 # Total de horas en una sola expresión
total_horas
````

<p>Si prefieres tener el total de horas como un entero, por supuesto puedes reemplazar la división de punto flotante con una división de enteros (es decir, <code>//</code>).</p>
